```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

# Sample Data (Replace this with your actual dataset)
data = {'age': [25, 34, 45, 23, 35, 42, 55, 65, 29, 30, 50, 48, 60, 33, 40],
        'income': [25000, 40000, 60000, 20000, 42000, 50000, 75000, 80000, 31000, 32000, 67000, 64000, 85000, 38000, 49000]}
df = pd.DataFrame(data)


df.dropna(inplace=True)
# Scaling features (K-Means is sensitive to feature scaling)
scaler = MinMaxScaler()
df[['age', 'income']] = scaler.fit_transform(df[['age', 'income']])


kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df[['age', 'income']])


df1 = df[df.cluster == 0]
df2 = df[df.cluster == 1]
df3 = df[df.cluster == 2]


plt.figure(figsize=(8, 6))
plt.scatter(df1.age, df1.income, color='red', label="Cluster 0")
plt.scatter(df2.age, df2.income, color='blue', label="Cluster 1")
plt.scatter(df3.age, df3.income, color='green', label="Cluster 2")


centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], color='black', marker='X', s=200, label="Centroids")


plt.xlabel("Age (scaled)")
plt.ylabel("Income (scaled)")
plt.legend()
plt.title("K-Means Clustering of Age vs. Income")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

file_path = '/content/income.csv'
df = pd.read_csv(file_path)

X = df[['Age', 'Income($)']]

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_predicted = kmeans.fit_predict(X)

df['cluster'] = y_predicted

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income($)'], c=df['cluster'], cmap='viridis', alpha=0.6, edgecolors='black')

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Income ($)', fontsize=12)
plt.title('K-Means Clustering: Age vs Income', fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

file_path = '/content/income.csv'
df = pd.read_csv(file_path)

X = df[['Age', 'Income($)']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

dbscan = DBSCAN(eps=0.5, min_samples=5)
y_predicted = dbscan.fit_predict(X_scaled)

df['cluster'] = y_predicted

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Income($)'], c=df['cluster'], cmap='viridis', alpha=0.6, edgecolors='black')

plt.xlabel('Age', fontsize=12)
plt.ylabel('Income ($)', fontsize=12)
plt.title('DBSCAN Clustering: Age vs Income', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("/content/Wholesale customers data.csv")

df_numeric = df.select_dtypes(include=[np.number])
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)

inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range, inertia, marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()

optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df_scaled)

plt.figure(figsize=(8,6))
plt.scatter(df_scaled[:, 0], df_scaled[:, 1], c=df['Cluster'], cmap='viridis', alpha=0.6)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering of Wholesale Customers")
plt.colorbar(label="Cluster")
plt.show()

df.head()
```
